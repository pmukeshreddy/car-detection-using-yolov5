```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Untitled folder/data/train_solution_bounding_boxes (1).csv")
```

```python
df.head()
```

```python
df["image"] = df["image"].apply(lambda x:x.split(".")[0])
df["x_center"] = (df["xmax"]+df["xmin"])/2
df["y_center"] = (df["ymax"]+df["ymin"])/2
df["w"] = df["xmax"] - df["xmin"]
df["h"] = df["ymax"] - df["ymin"]
img_h,img_w,num_channels = (380,676,3)
df["classes"] = 0
df["x_center"] = df["x_center"]/img_w
df["w"] = df["w"]/img_w
df["y_center"] = df["y_center"]/img_h
df["h"] = df["h"]/img_h
```
